```python
import pandas as pd
import mplfinance as mpf
import yfinance as yf

import matplotlib as plt
%matplotlib inline
```

```python
ticker = yf.Ticker("AMZN")

hist = ticker.history(period = "1y", interval = "1h")

mpf.plot(hist["2020-11-01":"2020-11-25"],
         volume = True, 
         type = "candle",
         mav = (20),
         tight_layout = True, 
         figratio = (20, 12), 
         style = "yahoo")
```
